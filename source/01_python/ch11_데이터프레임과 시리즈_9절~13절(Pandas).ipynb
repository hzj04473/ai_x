```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법


def year_month_append(df):
    '''
    df를 받아 연도와 월 추가후 return
    '''
    df = df.copy()
    # df.copy() 안하면 아래와 같이 에러 발생
    #     /var/folders/sr/fzxv95r54q53nl86rhm2c0zc0000gn/T/ipykernel_2806/3790721606.py:5: SettingWithCopyWarning:
    # A value is trying to be set on a copy of a slice from a DataFrame

    # See the caveats in the documentation: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
    #   df['연도'] = int(df['연도월'].split('년')[0])
    # /var/folders/sr/fzxv95r54q53nl86rhm2c0zc0000gn/T/ipykernel_2806/3790721606.py:5: SettingWithCopyWarning:
    # A value is trying to be set on a copy of a slice from a DataFrame

    # See the caveats in the documentation: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
    #   df['연도'] = int(df['연도월'].split('년')[0])
    
    
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1][:-1])
    return df
```

```python
import random
import numpy as np
import pandas as pd
import seaborn as sns
iris = sns.load_dataset('iris')
iris.shape
rowlen, collen = iris.shape
rowlen, collen
```

```python
# (1단계) iris 각 열마다 20개씩 인위적으로 격측치를 할당한다.
# 결측치가 각 열마다 20개씩 확인
print('임의의 20개 index 추출 : ', random.sample(range(rowlen), 20))
```

```python
# 결측치 할당
for col in range(collen - 1):
    iris.iloc[random.sample(range(rowlen), 20), col] = np.nan
```

```python
# 결측치 확인 - iris.isna() : 결측치인지 여부
# iris.info()
# iris.isna().sum()
print(iris.isnull().sum())
iris.tail()
```

```python
# 결측치가 하나라도 있는 행 출력
iris[iris['sepal_length'].isna() | iris['sepal_width'].isna()
     | iris['petal_length'].isna() | iris['petal_width'].isna()]
# iris[iris['sepal_length'].isna() | iris['sepal_width'].isna()
#      | iris['petal_length'].isna() | iris['petal_width'].isna()].index
```

```python
# (2-1단계) 결측치는 각 열의 평균으로 대체한다. : fillna 사용
# iris.median(numeric_only=True)
# iris.iloc[:, :-1].mean()
# iris
iris.fillna(iris.iloc[:, :-1].mean())
```

```python
iris.isna().sum()
```

```python
# (2-2단계) 결측치는 종별 열별 평균(groupby, pivot_table)로 대체한다. : fillna 사용x
iris_mean = iris.groupby(by=['species']).mean()
iris_mean
```

```python
#
iris.fillna(iris_mean)
```

```python
display(iris_mean)
display(iris_mean.loc['setosa','sepal_length'])
```

```python
# iris 한행을 매개변수로 받아 결츨치가 있는 경우 종별 열별 평균으로 대체하는 함수
def iris_fillna(irisrow):
    irisrow = irisrow.copy()
    for col, value in irisrow.items():
        # print(col, value)
        if isinstance(value, (int, float, np.number)) and np.isnan(value) :
            # print(irisrow[col], "이 결측치")
            irisrow[col] = iris_mean.loc[irisrow.species, col]
    return irisrow
```

```python
iris_fillna(iris.iloc[147])
```

```python
# apply 적용
iris = iris.apply(iris_fillna, axis=1)
```
