```python
# ★ 2차원 데이터 프레임의 apply 이용하는 방법


def year_month_append(df):
    '''
    df를 받아 연도와 월 추가후 return
    '''
    df = df.copy()
    # df.copy() 안하면 아래와 같이 에러 발생
    #     /var/folders/sr/fzxv95r54q53nl86rhm2c0zc0000gn/T/ipykernel_2806/3790721606.py:5: SettingWithCopyWarning:
    # A value is trying to be set on a copy of a slice from a DataFrame

    # See the caveats in the documentation: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
    #   df['연도'] = int(df['연도월'].split('년')[0])
    # /var/folders/sr/fzxv95r54q53nl86rhm2c0zc0000gn/T/ipykernel_2806/3790721606.py:5: SettingWithCopyWarning:
    # A value is trying to be set on a copy of a slice from a DataFrame

    # See the caveats in the documentation: https://pandas.pydata.org/pandas-docs/stable/user_guide/indexing.html#returning-a-view-versus-a-copy
    #   df['연도'] = int(df['연도월'].split('년')[0])
    
    
    df['연도'] = int(df['연도월'].split('년')[0])
    df['월'] = int(df['연도월'].split('년')[1][:-1])
    return df
```

```python
# year_month_append(new_df.iloc[0])
```

```python
new_df = new_df.apply(year_month_append, axis=1)
new_df
```

# 12절. 일괄변경하기 (결측치, 특정값)

## 12.1. fillna

- 대부분 결측치는 drop하거나, 평균값대체, 중위수대체, 그룹별 평균값, 그룹별 중위수, 이전값, 이후값 등으로 대체

```python
df = pd.DataFrame([
    [np.nan, 2, np.nan, 0],
    [3, 4, np.nan, 1],
    [np.nan, 3, np.nan, 5],
    [np.nan, 3, np.nan, 4],
],
                  columns=[list('ABCD')])

df
```

```python
# 모든 결측치를 0 대체
df.fillna(0)
```

```python
# 결측치를 이전행의 값으로 대체
display(df)
df.fillna(method="ffill")
```

```python
# 결측치를 다음행의 값으로 대체
display(df)
df.fillna(method='bfill')
```

```python
# 결측치를 열마다 특정값으로 대체
values = {'A': 85, 'B': 80, 'C': 75, 'D': 90}

df.fillna(value=values)
```

```python
# 결측치를 열별 평균치으로 대체
display(df)
display(df.mean(axis=0))
df.fillna(value=df.mean(axis=0))

```

```python
# 결측치를 열별 중위값으로 대체
display(df)
display(df.median(axis=0))
df.fillna(value=df.median(axis=0))
```

```python
# 결측치 대체를 2번쨰 행까지만 대체
display(df)
display(values)
df.fillna(value=values, limit=2)
```
