# 1. tensorflow v2

```python
import numpy as np
from tensorflow.keras.models import Sequential   # 모델객체생성
from tensorflow.keras.layers import Input, Dense # layer 층을 쌓을 떄 필요
```

```python
# 1. data set 확복
# 2. data 전처리
X_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_data = np.array([[0], [1], [1], [0]])
```

```python
# 3. 모델 구성
# layer층 x_data 2개 → 10개 → 20개 → 10개 → 1개
model = Sequential()
# model.add(Input(shape=(2,))) # 입력층
model.add(Dense(units=10, # 출력
                input_dim = 2, # 입력
                activation='relu')) # 활성화 함수

model.add(Dense(units=20,activation='relu'))
model.add(Dense(units=10,activation='relu'))
model.add(Dense(units=1,activation='sigmoid')) 
# 이진분류에서의 출력층 활성화 함수는 sigmoid
# 다중분류에서의 출력층 활성화 함수 softmax(출력합을 1로)

# 모델을 생성한 후에 꼭 summary를 해서 가능한지 확인
model.summary()
```

```python
# 4. 모델 학습과정 설정
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['binary_accuracy']) # 2진 분류
```

```python
# 5. 학습
hist = model.fit(
    X_data,
    y_data,  # 훈련데이터
    epochs=1200,  # 학습횟수
    # batch_size=2 # 개씩 자동으로
    verbose=2  # 학습로그출력 - 0: 출력무, 2: 상세출력
)
```

```python
# 5. 학습과정 살펴보기
hist.history.keys()
```

```python
import matplotlib.pyplot as plt
fig, loss_ax = plt.subplots(figsize=(10,5))
loss_ax.plot(hist.history['loss'],'r', label='loss')
loss_ax.set_xlabel('epochs')
loss_ax.set_ylabel('loss')
loss_ax.legend(loc='center right')

acc_ax = loss_ax.twinx() # loss_ax와 x축을 공유하는 acc_ax 생성
acc_ax.plot(hist.history['binary_accuracy'],'b', label='binary_accuracy')
acc_ax.set_ylabel('binary_accuracy')
acc_ax.legend(bbox_to_anchor=(0.99, 0.4))
```

```python
# 6. 학습평가하기

score = model.evaluate(X_data, y_data)
print('score :', score)

print('loss :', score[0])
print('binary_accuracy :', score[1])




```
