```python
# Q. ë‚˜ë¼ëª… â†’ (ì œì¼ ìœ ëª…í•œ ìŒì‹ â†’) ìŒì‹ì˜ ë ˆì‹œí”¼
# ë‹¤ìŒì˜ ìƒì„±í˜• aië¥¼ ë ì²´ì¸ìœ¼ë¡œ êµ¬í˜„í•˜ì‹œì˜¤.
# â–ª ì›í•˜ëŠ” ëª¨ë¸ì„ ì´ìš©í•˜ì—¬ ë‚˜ë¼ë¥¼ ì…ë ¥í•˜ë©´ í•´ë‹¹ ë‚˜ë¼ì˜ ê°€ì¥ ìœ ëª…í•œ ìŒì‹ì„ ì¶”ì²œí•˜ëŠ” ë ì²´ì¸ì„ êµ¬í˜„
# (food_chain))
# â–ª ìŒì‹ì„ ì…ë ¥í•˜ë©´ ë ˆì‹œí”¼ë¥¼ ìƒì„±í•˜ëŠ” ë ì²´ì¸ì„ êµ¬í˜„(recipe_chain)
# â–ª ìœ„ì˜ ë‘ ì²´ì¸ì„ ì—°ê²°í•˜ì—¬, ë‚˜ë¼ì´ë¦„ì„ ì…ë ¥ë°›ì•„, í•´ë‹¹ ë‚˜ë¼ì˜ ê°€ì¥ ìœ ëª…í•œ ìŒì‹ì˜ ë ˆì‹œí”¼ë¥¼ ìƒì„±í•˜ëŠ” ë ì²´ì¸ í”„ë¡œê·¸ë¨ì„ êµ¬í˜„í•˜ì‹œì˜¤.
```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

llm = ChatOllama(model="llama3.2:1b",temparature=0) # 0: ì¼ê´€ëœ ë‹µë³€

country_input = input("Enter the country name: ")

food_prompt = PromptTemplate(
    template = """What is the most famous food in {country_name}?
    Return only the name of the food.""",
    input_variables=["country_name"]
)

food_chain = food_prompt | llm | StrOutputParser()

food_result =food_chain.invoke({"country_name": country_input})
print(food_result)


recipe_prompt = PromptTemplate(
    template = """What is the recipe for {food_name}?
    Return only the recipe.""",
    input_variables=["food_name"]
)

recipe_chain = recipe_prompt | llm | StrOutputParser()

recipe_result = recipe_chain.invoke({"food_name": food_result})
print(recipe_result)
```

```python
# í´ë˜ìŠ¤ ê¸°ë°˜ êµ¬í˜„ (ì˜ë¬¸ ë³€ìˆ˜ëª…/í•¨ìˆ˜ëª… ì‚¬ìš©)
from langchain_ollama import ChatOllama
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

class CountryFoodRecipeGenerator:
    """ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥ë°›ì•„ ëŒ€í‘œ ìŒì‹ê³¼ ë ˆì‹œí”¼ë¥¼ ìƒì„±í•˜ëŠ” í´ë˜ìŠ¤"""
    
    def __init__(self, model_name="llama3.2:1b"):
        """LLM ëª¨ë¸ ì´ˆê¸°í™”"""
        try:
            self.llm = ChatOllama(model=model_name, temperature=0)
            self.output_parser = StrOutputParser()
            print(f"âœ… {model_name} ëª¨ë¸ì´ ì„±ê³µì ìœ¼ë¡œ ë¡œë“œë˜ì—ˆìŠµë‹ˆë‹¤.")
        except Exception as e:
            print(f"âŒ ëª¨ë¸ ë¡œë“œ ì‹¤íŒ¨: {e}")
            raise
    
    def create_food_chain(self):
        """ë‚˜ë¼ëª…ì„ ì…ë ¥ë°›ì•„ ëŒ€í‘œ ìŒì‹ì„ ì°¾ëŠ” ì²´ì¸ ìƒì„±"""
        food_prompt = PromptTemplate(
            template="""ë‹¤ìŒ ë‚˜ë¼ì˜ ê°€ì¥ ìœ ëª…í•œ ëŒ€í‘œ ìŒì‹ í•œ ê°€ì§€ë§Œ ì•Œë ¤ì£¼ì„¸ìš”: {country_name}
            
            ì¡°ê±´:
            - ìŒì‹ ì´ë¦„ë§Œ ê°„ë‹¨íˆ ë‹µë³€
            - ì„¤ëª…ì€ ì œì™¸í•˜ê³  ìŒì‹ëª…ë§Œ ë°˜í™˜
            """,
            input_variables=["country_name"]
        )
        return food_prompt | self.llm | self.output_parser
    
    def create_recipe_chain(self):
        """ìŒì‹ëª…ì„ ì…ë ¥ë°›ì•„ ë ˆì‹œí”¼ë¥¼ ìƒì„±í•˜ëŠ” ì²´ì¸ ìƒì„±"""
        recipe_prompt = PromptTemplate(
            template="""{food_name}ì˜ ê°„ë‹¨í•œ ë ˆì‹œí”¼ë¥¼ ì•Œë ¤ì£¼ì„¸ìš”.
            
            í˜•ì‹:
            - ì¬ë£Œ: (ì£¼ìš” ì¬ë£Œë§Œ ë‚˜ì—´)
            - ì¡°ë¦¬ë²•: (ê°„ë‹¨í•œ ë‹¨ê³„ë³„ ì„¤ëª…)
            - ì´ 20ì¤„ ì´ë‚´ë¡œ ê°„ê²°í•˜ê²Œ ì‘ì„±
            """,
            input_variables=["food_name"]
        )
        return recipe_prompt | self.llm | self.output_parser
    
    def validate_user_input(self, country_name):
        """ì‚¬ìš©ì ì…ë ¥ê°’ ìœ íš¨ì„± ê²€ì‚¬"""
        if not country_name or country_name.strip() == "":
            return False, "ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš”."
        
        if len(country_name.strip()) < 2:
            return False, "ì˜¬ë°”ë¥¸ ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš”."
        
        return True, "ìœ íš¨í•œ ì…ë ¥ì…ë‹ˆë‹¤."
    
    def execute_full_process(self, country_name):
        """ë‚˜ë¼ëª… â†’ ëŒ€í‘œìŒì‹ â†’ ë ˆì‹œí”¼ ì „ì²´ í”„ë¡œì„¸ìŠ¤ ì‹¤í–‰"""
        try:
            # 1. ì…ë ¥ê°’ ê²€ì¦
            is_valid, message = self.validate_user_input(country_name)
            if not is_valid:
                return f"âŒ {message}"
            
            # 2. ì²´ì¸ ìƒì„±
            food_chain = self.create_food_chain()
            recipe_chain = self.create_recipe_chain()
            
            # 3. ëŒ€í‘œ ìŒì‹ ì°¾ê¸°
            print(f"ğŸ” {country_name}ì˜ ëŒ€í‘œ ìŒì‹ì„ ì°¾ëŠ” ì¤‘...")
            famous_food = food_chain.invoke({"country_name": country_name.strip()})
            print(f"ğŸ½ï¸ ëŒ€í‘œ ìŒì‹: {famous_food}")
            
            # 4. ë ˆì‹œí”¼ ìƒì„±
            print(f"ğŸ‘¨â€ğŸ³ {famous_food} ë ˆì‹œí”¼ë¥¼ ìƒì„±í•˜ëŠ” ì¤‘...")
            recipe = recipe_chain.invoke({"food_name": famous_food})
            
            # 5. ìµœì¢… ê²°ê³¼ í¬ë§¤íŒ…
            final_result = f"""
ğŸŒ ë‚˜ë¼: {country_name}
ğŸ½ï¸ ëŒ€í‘œ ìŒì‹: {famous_food}

ğŸ“‹ ë ˆì‹œí”¼:
{recipe}
            """
            return final_result.strip()
            
        except Exception as e:
            return f"âŒ ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}"

# ì‹¤ì œ ì‚¬ìš© ì˜ˆì‹œ
if __name__ == "__main__":
    # ìƒì„±ê¸° ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
    recipe_generator = CountryFoodRecipeGenerator()
    
    # ì‚¬ìš©ì ì…ë ¥ ë°›ê¸°
    country_input = input("ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš” (ì˜ˆ: í•œêµ­, ì´íƒˆë¦¬ì•„, ì¼ë³¸): ")
    
    # ì „ì²´ í”„ë¡œì„¸ìŠ¤ ì‹¤í–‰
    result = recipe_generator.execute_full_process(country_input)
    print("\n" + "="*50)
    print(result)
    print("="*50)

```

```python
# LCELì„ í™œìš©í•œ ë” ê°„ë‹¨í•œ ë²„ì „ (ì˜ë¬¸ ë³€ìˆ˜ëª…/í•¨ìˆ˜ëª… ì‚¬ìš©)
from langchain_core.runnables import RunnablePassthrough

def create_lcel_country_food_recipe():
    """LCELì„ í™œìš©í•œ ê°„ê²°í•œ êµ¬í˜„"""
    
    # LLM ì´ˆê¸°í™”
    llm = ChatOllama(model="llama3.2:1b", temperature=0)
    output_parser = StrOutputParser()
    
    # 1ë‹¨ê³„: ë‚˜ë¼ â†’ ëŒ€í‘œìŒì‹ ì²´ì¸
    food_prompt = PromptTemplate(
        template="ë‹¤ìŒ ë‚˜ë¼ì˜ ê°€ì¥ ìœ ëª…í•œ ëŒ€í‘œ ìŒì‹ í•œ ê°€ì§€ë§Œ ì•Œë ¤ì£¼ì„¸ìš”: {country_name}\nìŒì‹ ì´ë¦„ë§Œ ê°„ë‹¨íˆ ë‹µë³€í•´ì£¼ì„¸ìš”.",
        input_variables=["country_name"]
    )
    food_chain = food_prompt | llm | output_parser
    
    # 2ë‹¨ê³„: ìŒì‹ â†’ ë ˆì‹œí”¼ ì²´ì¸  
    recipe_prompt = PromptTemplate(
        template="""{food_name}ì˜ ê°„ë‹¨í•œ ë ˆì‹œí”¼ë¥¼ ì•Œë ¤ì£¼ì„¸ìš”.
        
        í˜•ì‹:
        - ì¬ë£Œ: (ì£¼ìš” ì¬ë£Œë§Œ)
        - ì¡°ë¦¬ë²•: (ê°„ë‹¨í•œ ë‹¨ê³„ë³„ ì„¤ëª…)
        - 20ì¤„ ì´ë‚´ë¡œ ê°„ê²°í•˜ê²Œ
        """,
        input_variables=["food_name"]
    )
    
    # 3ë‹¨ê³„: í†µí•© ì²´ì¸ êµ¬ì„± (LCELì˜ í•µì‹¬!)
    integrated_chain = (
        {"country_name": RunnablePassthrough()}  # ì…ë ¥ ê·¸ëŒ€ë¡œ ì „ë‹¬
        | {"food_name": food_chain}              # ë‚˜ë¼ëª… â†’ ìŒì‹ëª… ë³€í™˜
        | recipe_prompt                          # ìŒì‹ëª…ì„ ë ˆì‹œí”¼ í”„ë¡¬í”„íŠ¸ì— ì ìš©
        | llm                                    # LLM ì‹¤í–‰
        | output_parser                          # ë¬¸ìì—´ë¡œ ë³€í™˜
    )
    
    return integrated_chain

# ì‹¤í–‰ ì˜ˆì‹œ
if __name__ == "__main__":
    print("ğŸš€ LCEL ë°©ì‹ìœ¼ë¡œ ë‚˜ë¼ìŒì‹ë ˆì‹œí”¼ ìƒì„±ê¸° ì‹œì‘")
    
    try:
        chain = create_lcel_country_food_recipe()
        country_input = input("ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
        
        if country_input.strip():
            print(f"\nğŸ”„ {country_input} ì²˜ë¦¬ ì¤‘...")
            result = chain.invoke(country_input.strip())
            print(f"\nğŸ“‹ ê²°ê³¼:\n{result}")
        else:
            print("âŒ ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš”.")
            
    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
```

```python
# ê°„ë‹¨í•œ í•¨ìˆ˜í˜• ì ‘ê·¼ë²• (ì›ë³¸ ì½”ë“œ ìˆ˜ì • ë²„ì „, ì˜ë¬¸ ë³€ìˆ˜ëª…/í•¨ìˆ˜ëª… ì‚¬ìš©)
def simple_country_food_recipe(country_name):
    """ì›ë³¸ ì½”ë“œì˜ ë¬¸ì œì ì„ ìˆ˜ì •í•œ ê°„ë‹¨í•œ ë²„ì „"""
    
    # LLM ì´ˆê¸°í™”
    llm = ChatOllama(model="llama3.2:1b", temperature=0)  # temperature ì˜¤íƒ€ ìˆ˜ì •
    output_parser = StrOutputParser()
    
    # 1ë‹¨ê³„: ëŒ€í‘œ ìŒì‹ ì°¾ê¸°
    food_prompt = PromptTemplate(
        template="ë‹¤ìŒ ë‚˜ë¼ì˜ ê°€ì¥ ìœ ëª…í•œ ëŒ€í‘œ ìŒì‹ í•œ ê°€ì§€ë§Œ ì•Œë ¤ì£¼ì„¸ìš”: {country_name}\nìŒì‹ ì´ë¦„ë§Œ ê°„ë‹¨íˆ ë‹µë³€í•´ì£¼ì„¸ìš”.",
        input_variables=["country_name"]
    )
    food_chain = food_prompt | llm | output_parser
    famous_food = food_chain.invoke({"country_name": country_name})
    
    print(f"ğŸ½ï¸ {country_name}ì˜ ëŒ€í‘œ ìŒì‹: {famous_food}")
    
    # 2ë‹¨ê³„: ë ˆì‹œí”¼ ìƒì„±
    recipe_prompt = PromptTemplate(
        template="""{food_name}ì˜ ê°„ë‹¨í•œ ë ˆì‹œí”¼ë¥¼ ì•Œë ¤ì£¼ì„¸ìš”.
        
        í˜•ì‹:
        - ì¬ë£Œ: (ì£¼ìš” ì¬ë£Œë§Œ ë‚˜ì—´)
        - ì¡°ë¦¬ë²•: (ê°„ë‹¨í•œ ë‹¨ê³„ë³„ ì„¤ëª…)  
        - ì´ 20ì¤„ ì´ë‚´ë¡œ ê°„ê²°í•˜ê²Œ ì‘ì„±
        """,
        input_variables=["food_name"]
    )
    recipe_chain = recipe_prompt | llm | output_parser
    recipe = recipe_chain.invoke({"food_name": famous_food})
    
    # 3ë‹¨ê³„: ê²°ê³¼ í¬ë§¤íŒ…
    final_result = f"""
ğŸŒ ë‚˜ë¼: {country_name}
ğŸ½ï¸ ëŒ€í‘œ ìŒì‹: {famous_food}

ğŸ“‹ ë ˆì‹œí”¼:
{recipe}
    """
    
    return final_result.strip()

# ì‹¤í–‰ ì˜ˆì‹œ
if __name__ == "__main__":
    try:
        country_input = input("ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
        
        if country_input.strip():
            result = simple_country_food_recipe(country_input.strip())
            print("\n" + "="*50)
            print(result)
            print("="*50)
        else:
            print("âŒ ë‚˜ë¼ ì´ë¦„ì„ ì…ë ¥í•´ì£¼ì„¸ìš”.")
            
    except Exception as e:
        print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
```
